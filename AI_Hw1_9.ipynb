```python
import numpy as np

class GradientDescentOptimizer:

    def __init__(self, learning_rates, steps, trials, bounds):
        self.learning_rates = learning_rates
        self.steps = steps
        self.trials = trials
        self.bounds = bounds

    def objective_function(self, x, y):
        return (5 * x * x) + (40 * x) + (y * y) - (12 * y) + 127

    def gradient(self, x, y):
        return np.array([(10 * x) + 40, (2 * y) - 12])

    def gradient_descent(self, x0, learning_rate):
        current_position = np.array(x0, dtype=float)
        for _ in range(self.steps):
            grad = self.gradient(current_position[0], current_position[1])
            current_position = current_position - learning_rate * grad
        return current_position

    def find_best_result(self, learning_rate):
        best_position = None
        for _ in range(self.trials):
            x0 = np.random.uniform(self.bounds[0], self.bounds[1], size=(2,))
            result = self.gradient_descent(x0, learning_rate)

            if (best_position is None or
                self.objective_function(result[0], result[1]) < self.objective_function(best_position[0], best_position[1])):
                best_position = result

        return best_position

    def run_experiment(self):
        for experiment in range(3):
            print(f"=== Experiment Set {experiment + 1} ===")
            print(f"{'eta':<10} | {'Best Position (x, y)':<25} | {'f(x, y)':<15}")
            print("-" * 55)
            for learning_rate in self.learning_rates:
                best_result = self.find_best_result(learning_rate)
                f_val = self.objective_function(best_result[0], best_result[1])
                print(f"{learning_rate:<10} | ({best_result[0]:.6f}, {best_result[1]:.6f}){'':<5} | {f_val:.6f}")
            print()


if __name__ == "__main__":
    LEARNING_RATES = [0.1, 0.01, 0.001]
    STEPS = 500
    TRIALS = 10
    BOUNDS = [-10, 10]

    optimizer = GradientDescentOptimizer(LEARNING_RATES, STEPS, TRIALS, BOUNDS)
    optimizer.run_experiment()
```
